```python
# Preprocessing and augmentation
def showDataDist(train):
    wedge = [train['label'].value_counts()[0],train['label'].value_counts()[1],
             train['label'].value_counts()[2],train['label'].value_counts()[3],
             train['label'].value_counts()[4],train['label'].value_counts()[5],
             train['label'].value_counts()[6]]

    perc = [train['label'].value_counts()[0]/len(train),
            train['label'].value_counts()[1]/len(train),
            train['label'].value_counts()[2]/len(train),
            train['label'].value_counts()[3]/len(train),
            train['label'].value_counts()[4]/len(train),
               train['label'].value_counts()[5]/len(train),
            train['label'].value_counts()[6]/len(train)
           ]
    plt.pie(wedge,labels=['Passenger ship   - '+ format(perc[0]*100, '.2f') + '%','Military ship  - '+ format(perc[1]*100, '.2f') + '%','Container ship  - '+ format(perc[2]*100, '.2f') + '%','Tanker - '+ format(perc[3]*100, '.2f') + '%','Tug - '+ format(perc[4]*100, '.2f') + '%'
                         + '%','Sailing vessel - '+ format(perc[4]*100, '.2f') + '%'+'%','Fishing vessels  - '+ format(perc[4]*100, '.2f') + '%'],
            shadow=True,radius = 2.0)
```

```python
#Model Building

def buildModel(NUM_CLASSES):
    base_model =MobileNet(weights='imagenet',include_top=True)

    x=base_model.layers[-2].output
    # x=GlobalAveragePooling2D()(x)
    preds=Dense(NUM_CLASSES,activation='softmax')(x)
    model=Model(inputs=base_model.input,outputs=preds)

    for layer in model.layers:
        layer.trainable=False
    for layer in model.layers[-1:]:
        layer.trainable=True
    model.summary()
    return model

def startTraining(model,train_df,val_df,BATCH_SIZE=64,epochs=50,checkpoints_path='snapshots2'):
    train_datagen = ImageDataGenerator(
        preprocessing_function=preprocess_input)
    val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
    
    train_generator = train_datagen.flow_from_dataframe(
        dataframe=train_df,
        x_col="filename",
        y_col="label",
        target_size=(224, 224),
        batch_size=BATCH_SIZE,
        class_mode='categorical')
    validation_generator = val_datagen.flow_from_dataframe(
            dataframe=val_df,
            x_col="filename",
            y_col="label",
            target_size=(224, 224),
            batch_size=BATCH_SIZE,
            class_mode='categorical')
    
    model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
    
    #
    checkpoint_path = checkpoint_path+"/cp-{epoch:04d}.ckpt"
    checkpoint_dir = os.path.dirname(checkpoint_path)
    cp_callback = keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    save_weights_only=True,
    period=3)


    history=model.fit_generator(
        train_generator,
        steps_per_epoch=int(train_df.shape[0]/(BATCH_SIZE)),
        epochs=epochs,
        callbacks=[cp_callback],
        validation_data=validation_generator,
        validation_steps=int(val_df.shape[0]/(BATCH_SIZE))
    )

    print(history.history.keys())
    return history
```

```python
model=buildModel(7);
print(train_aug_df.shape,val.shape)
history=startTraining(model,train_aug_df,val,epochs=5)
```
